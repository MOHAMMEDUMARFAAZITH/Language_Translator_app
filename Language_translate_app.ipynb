```python
import docx
from googletrans import Translator

# Replace this with the path to your DOCX file
docx_file_path = r"C:\Users\RAZITH\Downloads\aws-skill-builder.docx"

doc = docx.Document(docx_file_path)

extracted_text = ""
for paragraph in doc.paragraphs:
    extracted_text += paragraph.text + "\n"

# Split the extracted text into smaller chunks for translation
chunk_size = 500  # Adjust as needed
chunks = [extracted_text[i:i + chunk_size] for i in range(0, len(extracted_text), chunk_size)]

translator = Translator()
translated_chunks = []

for chunk in chunks:
    translation = translator.translate(chunk, dest="Fr")
    translated_chunks.append(translation.text)

translated_text = "\n".join(translated_chunks)
print(translated_text)
```
